### Dungeon Egg Hunt

Using object-oriented programming principles, create an egg hunt game. Your objective is to go through the dungeon to collect the basket and the eggs. Once you've retrieved all of the items, exit through the door.

**RULES**:

`1. ` Player, Monster, Door, Basket and the 3 Eggs must all start at a random locations on a grid.

`2. ` Player must collect the Basket first before collecting the 3 Eggs. Player will not be able to collect any Eggs until it first has collected the Basket.

`3. ` Once Player collects the Basket, then and only then will it be able to collect any Eggs.

`4. ` When Player collects the Basket and the 3 Eggs, Player will need to head toward the Door.

`5. ` The entire time the Player is searching for the Eggs, Basket and the Door, there will be a Monster trying to eat the Player.

`6. ` If the Monster lands on the same space as the Player, or the Player lands on the same space as the Monster, the game ends and the game will prompt the user to play again.

`7. ` If Player reaches the Door before the Monster catches them, you win the game and the game will prompt the user to play again.

`8. ` The Player does not know where any of the other tokens are, so all of the tokens with the exception of the Player will be hidden.

**Note**: For debugging purposes, you can create tokens for each of the other tokens. For example:<br>
* Player (P)
* Monster (M)
* Eggs (O)
* Basket (U)
* Door (D)

```python
import random
class Game:
    def __init__(self):
        self.coordinate = [
            (0,0),(1,0),(2,0),(3,0),(4,0),
            (0,1),(1,1),(2,1),(3,1),(4,1),
            (0,2),(1,2),(2,2),(3,2),(4,2),
            (0,3),(1,3),(2,3),(3,3),(4,3),
            (0,4),(1,4),(2,4),(3,4),(4,4),
        ]
        self.coordinates = [list(ele) for ele in self.coordinate]
        self.player=Player()
        self.monster=Monster()
        self.egg1=Egg()
        self.egg2=Egg()
        self.egg3=Egg()
        self.basket=Basket()
        self.door=Door()
    def draw_map(self):
        field = []
        field_rows = []
        field = self.coordinates
        field_rows = [field[:5], field[5:10], field[10:15], field[15:20], field[20:25]]
        print(f'monster loc on draw {self.monster.loc}Player loc {self.player.loc}')
        for i in range(len(field_rows)):
            for cell_index in range(len(field_rows[i])):
                if (self.monster.loc[0] == field_rows[i][cell_index][0]) and (self.monster.loc[1] == field_rows[i][cell_index][1]):
                    field_rows[i][cell_index] = "[M]"
                elif (self.player.loc[0] == field_rows[i][cell_index][0]) and (self.player.loc[1] == field_rows[i][cell_index][1]):
                    field_rows[i][cell_index] = "[P]"
                elif (self.egg1.loc[0] == field_rows[i][cell_index][0]) and (self.egg1.loc[1] == field_rows[i][cell_index][1]):
                    field_rows[i][cell_index] = "[E]"
                elif (self.egg2.loc[0] == field_rows[i][cell_index][0]) and (self.egg2.loc[1] == field_rows[i][cell_index][1]):
                    field_rows[i][cell_index] = "[E]"
                elif (self.egg3.loc[0] == field_rows[i][cell_index][0]) and (self.egg3.loc[1] == field_rows[i][cell_index][1]):
                    field_rows[i][cell_index] = "[E]"
                elif (self.basket.loc[0] == field_rows[i][cell_index][0]) and (self.basket.loc[1] == field_rows[i][cell_index][1]):
                    field_rows[i][cell_index] = "[B]"
                elif (self.door.loc[0] == field_rows[i][cell_index][0]) and (self.door.loc[1] == field_rows[i][cell_index][1]):
                    field_rows[i][cell_index] = "[D]"
                else:
                    field_rows[i][cell_index] = "[ ]"
            print(''.join(field_rows[i]))
    def main(self):
        inventory = []
        input("""Welcome to Dungeon Game
        Press Enter to start the game""")
        while True:
            self.draw_map()
            print(f'basket loc{self.basket.loc}')
            print(f'player loc{self.player.loc} pre move')
            self.player.move()
            print(f'player loc{self.player.loc} post move')
            if self.monster.loc == self.player.loc:
                print("""
                You have encountered a monster!
                Game Over.
                """)
                break
            if self.egg1.loc == self.player.loc:
                if "B" in inventory:
                    inventory.append("E")
                if "B" not in inventory:
                    continue
            if self.egg2.loc == self.player.loc:
                if "B" in inventory:
                    inventory.append("E")
                if "B" not in inventory:
                    continue
            if self.egg3.loc == self.player.loc:
                if "B" in inventory:
                    self.player.append("E")
                if "B" not in inventory:
                    continue
            if self.basket.loc == self.player.loc:
                inventory.append("B")
            if self.door.loc == self.player.loc:
                if "B" and "E" and "E" and "E" in inventory:
                    print("Yay! You Won!")
                    break
            print(f'moster loc{self.monster.loc} pre move')
            self.monster.move()
            print(f'moster loc{self.monster.loc} post move')
class Token:
    def __init__(self):
        self.loc = random.choice(game.coordinates)
class Player(Token):
    def __init__(self):
        super().__init__()
    def move(self):
        user_input=input("Which way? (n, e, s, w)")
        if user_input.lower() == "n":
            self.loc[1] -= 1
            if self.loc[1]<0:
                self.loc[1]=0
        if user_input.lower() == "s":
            self.loc[1] += 1
            if self.loc[1]>4:
                self.loc[1] = 4
        if user_input.lower() == "e":
            self.loc[0] += 1 
            if self.loc[0]>4:
                self.loc[0]=4
        if user_input.lower() == "w":
            self.loc[0] -= 1
            if self.loc[0]<0:
                self.loc[0]=0
class Monster(Token):
    def __init__(self):
        super().__init__()
    def move(self):
        self.loc[1] = random.randint(0, 4)
        self.loc[0] = random.randint(0, 4)
        if self.loc[0]>4 or self.loc[1]>4 or self.loc[0]<0 or self.loc[0]<0:
            self.move()
class Egg(Token):
    def __init__(self):
        super().__init__()
class Basket(Token):
    def __init__(self):
        super().__init__()
class Door(Token):
    def __init__(self):
        super().__init__()
game=Game()               
game.main()
```
